```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.xlim(11,-2)
plt.ylim(2.2,-1.3)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
plt.axis("equal")
```

```python
#去掉边界的空白
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
plt.axis("tight")
```

```python
#在同一个坐标上画两个图
```

```python
plt.plot(x,np.sin(x),label="sin(x)")
plt.plot(x,np.cos(x),label="cos(x)")
plt.axis("equal")
plt.legend()
```

```python
import matplotlib.pyplot as plt
plt.style.use("classic")
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
mg=np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(mg.randn(500,6),0)
```

```python
plt.plot(x,y)
plt.legend("abcded",ncol=2,loc="upper left")
```

```python
import seaborn as sns
sns.set()
plt.plot(x,y)
plt.legend("abcdef",ncol=2,loc="upper left")
```
